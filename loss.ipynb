```python
import matplotlib.pyplot as plt
import pandas as pd


def plot_loss(ax, title, df, gen_color='blue', disc_color='red'):

    ax.set_xlabel('Epochs')
    ax.set_ylabel('Loss')
    ax.set_title(title)
    df.generator_loss.plot(label='Generator loss', color=gen_color)
    df.discriminator_loss.plot(label='Discriminator loss', color=disc_color)
    plt.legend()


fig = plt.figure(figsize=(60, 16))
cols = 5

ax1 = fig.add_subplot(2, cols, 1)
df1 = pd.read_csv('../persistent/model/2022_08_04_11_08_31_loss_output_10000.txt', sep=',', header=0).iloc[::100]
plot_loss(ax1, 'Onto-DP-CGAN 1e-4 lr 5 critic steps 10000 epochs', df1)

ax2 = fig.add_subplot(2, cols, 4)
df2 = pd.read_csv('../persistent/model/2022_08_05_07_24_28_loss_output_20000.txt', sep=',', header=0).iloc[::100]
plot_loss(ax2, 'Onto-DP-CGAN 2e-4 lr 1 critic step 10000 epochs', df2)

ax3 = fig.add_subplot(2, cols, 2)
df3 = pd.read_csv('../persistent/model/2022_08_04_15_13_33_loss_output_1000.txt', sep=',', header=0).iloc[::10]
plot_loss(ax3, 'Onto-DP-CGAN 2e-4 lr 1 critic step 1000 epochs', df3)

ax4 = fig.add_subplot(2, cols, 5)
df4 = pd.read_csv('../test_loss/dp_cgans/ontodata_loss_output_10000.txt', sep=',', header=0).iloc[::100]
plot_loss(ax4, 'DP-CGAN 2e-4 lr 1 critic step 10000 epochs', df4)

ax5 = fig.add_subplot(2, cols, 3)
df5 = pd.read_csv('../test_loss/dp_cgans/loss_output_1000.txt', sep=',', header=0).iloc[::10]
plot_loss(ax5, 'DP-CGAN 2e-4 lr 1 critic step 1000 epochs', df5)

ax6 = fig.add_subplot(2, cols, 6)
df6 = pd.read_csv('../persistent/model/2022_08_05_09_38_11_loss_output_4000.txt', sep=',', header=0).iloc[::100]
plot_loss(ax6, 'Onto-DP-CGAN 2e-4 lr 1 critic step 4000 epochs', df6)

ax7 = fig.add_subplot(2, cols, 7)
df7 = pd.read_csv('../persistent/model/2022_08_08_08_13_37_loss_output_10000.txt', sep=',', header=0).iloc[::100]
plot_loss(ax7, 'Onto-DP-CGAN 2e-4 lr 1 critic step 10000 epochs ONLY RD EMBEDDING', df7)

ax8 = fig.add_subplot(2, cols, 8)
df8 = pd.read_csv('../persistent/model/2022_08_09_07_08_54_loss_output_10000.txt', sep=',', header=0).iloc[::100]
plot_loss(ax8, 'Onto-DP-CGAN 2e-4 lr 1 critic step 10000 epochs (fix) ONLY RD EMBEDDING', df8, gen_color='green')

ax9 = fig.add_subplot(2, cols, 9)
df9 = pd.read_csv('../persistent/model/2022_08_08_15_16_40_loss_output_10000.txt', sep=',', header=0).iloc[::100]
plot_loss(ax9, 'Onto-DP-CGAN 2e-4 lr 1 critic step 10000 epochs (fix) all embeddings', df9, gen_color='green')
```

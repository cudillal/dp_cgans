```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../persistent/model/2022_08_04_11_08_31_loss_output_10000.txt', sep=',', header=0)
df = df.iloc[::100]

fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.set_title('GAN Loss')
df.generator_loss.plot(label='Generator loss', color='green')
df.discriminator_loss.plot(label='Discriminator loss', color='red')
plt.legend()
```
